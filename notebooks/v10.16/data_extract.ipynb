## EACH Champion Winning Rate(table1)

```
# seperate to two team
newtable1 = df[['t1p1_champId', 't1p2_champId', 't1p3_champId', 't1p4_champId', 't1p5_champId', 't1_win']]
newtable2 = df[['t2p1_champId', 't2p2_champId', 't2p3_champId', 't2p4_champId', 't2p5_champId', 't1_win']]

newtable2['t1_win'] = newtable2['t1_win'].replace({0:1, 1:0})
# count winrate for team1
melttable = pd.melt(newtable1,id_vars=['t1_win'], var_name='player', value_name= 'champId')
melttable = melttable.dropna(subset=['t1_win'])
# count pick number
melttable['champId'] = melttable['champId'].astype(int)
rolecount = melttable.groupby('champId').count()['t1_win']
# compute win count and win rate
wincount = melttable.groupby('champId').sum()['t1_win']
winpercent = wincount / rolecount
# final table for team1
result1 = pd.concat([rolecount, wincount, winpercent], axis = 1)
result1.columns = ['count', 'win', 'winpercent']

# same for table 2
melttable2 = pd.melt(newtable2,id_vars=['t1_win'], var_name='player', value_name= 'champId')

rolecount2 = melttable2.groupby('champId').count()['t1_win']
wincount2 = melttable2.groupby('champId').sum()['t1_win']
winpercent2 = wincount2 / rolecount2
result2 = pd.concat([rolecount2, wincount2, winpercent2],axis= 1)
result2.columns = ['count', 'win', 'winpercent']
# converge to one table
rolecount_total = rolecount + rolecount2
wincount_total = wincount + wincount2
winpercent_total = wincount_total / rolecount_total

result = pd.concat([rolecount_total, wincount_total, winpercent_total], axis = 1)
result.columns = ['count', 'win', 'winpercent']

result.to_csv('table1_cw.csv', index = True)
```

## Champion in a different position(table3)

```
positions = ['SUPPORT', 'ADC', 'MIDDLE', 'JUNGLE', 'TOP']


# Team 1
t1_fields = []

for position in range(1, 6):
    field = f't1p{position}_champId'
    t1_fields.append(field)

t1 = df[t1_fields].copy()
t1['t1_win'] = df['t1_win']
t1 = t1.melt(id_vars=['t1_win'], value_vars=t1_fields[:], var_name='position', value_name='champId')
t1['position'] = t1['position'].replace(dict(zip(t1_fields, positions)))
t1 = t1.groupby(['position', 'champId']).agg({'t1_win': ['count', 'sum', 'mean']})
t1.columns = ['appearances', 'wins', 'winrate']
t1 = t1.reset_index()


# Team 2
t2_fields = []

for position in range(1, 6):
    field = f't2p{position}_champId'
    t2_fields.append(field)

t2 = df[t2_fields].copy()
t2['t2_win'] = df['t1_win'].replace({0: 1, 1: 0})
t2 = t2.melt(id_vars=['t2_win'], value_vars=t2_fields[:], var_name='position', value_name='champId')
t2['position'] = t2['position'].replace(dict(zip(t2_fields, positions)))
t2 = t2.groupby(['position', 'champId']).agg({'t2_win': ['count', 'sum', 'mean']})
t2.columns = ['appearances', 'wins', 'winrate']
t2 = t2.reset_index()
```

```
# Merge
merged_table = pd.merge(t1, t2, on=['position', 'champId'], how='outer')
merged_table = merged_table.fillna(0)
merged_table['appearances'] = merged_table['appearances_x'].add(merged_table['appearances_y'])
merged_table['wins'] = merged_table['wins_x'].add(merged_table['wins_y'])
merged_table = merged_table.drop(columns=['appearances_x', 'appearances_y', 'wins_x', 'wins_y', 'winrate_x', 'winrate_y'])
merged_table['winrate'] = merged_table['wins'] / merged_table['appearances']
pivot_table = merged_table.pivot_table(index='champId', 
                                        columns='position', 
                                        values=['appearances', 'wins', 'winrate'],
                                        aggfunc='sum')
pivot_table.columns = ['_'.join(reversed(col)).strip() for col in pivot_table.columns.values]
pivot_table = pivot_table.sort_index(axis=1)
pivot_table = pivot_table.reset_index()
pivot_table.to_csv('table3.csv', index = False)
```
